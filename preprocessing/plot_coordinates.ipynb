```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#data = pd.read_csv('2000_coordinates.csv', header=None)
#data = pd.read_csv('2000_coordinates.csv', header=None)
data = pd.read_csv('5_coordinates.csv', header=None)
```

```python
#data['id'] = data.index % 2000
data['id'] = data.index % 5
```

```python
data.head()
```

```python
plt.gray()

fig, ax = plt.subplots(figsize=(20, 20))
for i in range(5):
    ax.plot(data[data['id'] == i][0], data[data['id'] == i][1], '-', color=[.5,.5,.5], linewidth=2)
    last_point = data[data['id'] == i].iloc[-1]
    ax.plot(last_point[0], last_point[1], 'o', color=[.2,.2,.2], alpha=.8)
ax.axis('equal')
ax.axis('off')
plt.savefig("5_coordinates.pdf")
plt.show()
```

## Time evolution of 2,000 points

Five closeby points traverse space from top left to bottom right. Due to their sensitivity on the inital positions their movement evolves chaotically.
